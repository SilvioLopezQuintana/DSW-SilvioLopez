#####Las variables  que se consideran relevantes al modelo, son las columnas correspondientes a los sintomas y la severidad de estas .

```python
datos=datos_cleaned.iloc[:,[0,1,2,3,4,6,7,8,9,22]]
datos
```

######Debido a que no hay informaci√≥n de los casos positivos de covid en el dataframe, se asume como casos positivos cuando los sintomas son severos, es decir Severity_Severe =1


#####Se visulizan los casos severos (contagiados) y no severos (no cantagiados)

```python
total_personas = len(datos)
porcentaje_no = (datos['Severity_Severe'] == 0).sum() / total_personas * 100
porcentaje_si = (datos['Severity_Severe'] == 1).sum() / total_personas * 100

# Crear un gr√°fico de barras
categorias = ['No', 'Si']
porcentajes = [porcentaje_no, porcentaje_si]

plt.figure(figsize=(4, 3))
plt.bar(categorias, porcentajes, color=['red', 'green'])
plt.title('Contagiados')
plt.ylabel('Porcentaje (%)')

# Mostrar el gr√°fico
plt.show()
```

##Modelado con un √°rbol de decisi√≥n, utilizando busqueda de rejilla con validaci√≥n cruzada

```python
x=datos.iloc[:,0:8].values
x
```

```python
x1=datos.iloc[:,0:8]
x1
```

```python
y=datos.iloc[:,-1].values
y
```

```python
y1=datos[['Severity_Severe']]
y1
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import f1_score,accuracy_score
modelo1=DecisionTreeClassifier()
from sklearn.model_selection import RandomizedSearchCV
parametros={'criterion':['gini','entropy'],
             'max_depth':[4,5,6]}
```

```python
metrica=['f1_macro','accuracy']
```

```python
rejilla1=RandomizedSearchCV(modelo1,parametros,scoring=metrica,refit='f1_macro',cv=5,n_iter=3)
```

```python
rejilla1.fit(x,y)
```

```python
mejor_score1= rejilla1.best_score_*100
mejores_parametros1=rejilla1.best_params_
```

```python
print(f' f1_macro fue : {mejor_score1 :.2f} %')
print(f'Los mejores hiper par√°metros aleatoreos fueron : {mejores_parametros1}')
```

```python
#Tabla de iterciones del arbol de decisis√≥n
resultados1 =pd.DataFrame(rejilla1.cv_results_)[['params','mean_test_f1_macro','mean_test_accuracy']]
resultados1
```

```python
exactitud1=resultados1.iloc[0,2]*100
print(f'La exactitud del modelo fue : {exactitud1} %')
```

##### Con una exactitud del 75 % y un f1 macro de 42.86 % , se concluye que no es un modelo muy confiable

##Visualizaci√≥n del √°rbol

```python
from sklearn.tree import export_graphviz
import graphviz
```

```python
mejor_arbol = rejilla1.best_estimator_

# Visualizar el √°rbol de decisi√≥n
dato_arbol = export_graphviz(
    mejor_arbol,
    out_file=None,
    feature_names=x1.columns,
    class_names=y1.columns,
    filled=True,
    rounded=True,
    special_characters=True

)

graph = graphviz.Source(dato_arbol)
```

##Modelado con el algoritmo knn (vecino m√°s cercano)


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
knn=KNeighborsClassifier()
```

```python
parametros2={'n_neighbors' :[1,3],
              'p':[1,2]
             }
```

```python
rejilla2=RandomizedSearchCV(knn,parametros2,scoring=metrica,refit='f1_macro',cv=5,n_iter=3)
```

```python
rejilla2.fit(x,y)
```

```python
mejor_score2= rejilla2.best_score_*100
mejores_parametros2=rejilla2.best_params_
```

```python
print(f' f1_macro de knn fue : {mejor_score2 :.2f} %')
print(f'Los mejores hiper par√°metros aleatoreos de knn fueron : {mejores_parametros2}')
```

```python
#Tabla de ieraciones de knn
resultados2 =pd.DataFrame(rejilla2.cv_results_)[['params','mean_test_f1_macro','mean_test_accuracy']]
resultados2
```

```python
exactitud2=resultados2.iloc[1,2]*100
print(f'La exactitud del modelo knn fue : {exactitud2 :.2f} %')
```

#####La primera iteraci√≥n obtuvo una exactitud del 73 % y un f1_macro del 45%.
#####La segunda iteraci√≥n obtuvo una exactitud del 69% y un f1_macro del 48%.
#####Aunque no hay diferencias muy significativas entre la exactitud y el
#####f1_macro de las dos replicas, se opta por elegir como mejor modelo a la

#Modelado con nave bayes, utilizando busqueda de rejilla y validaci√≥n cruzada

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
```

```python
bayes=GaussianNB()
```

```python
parametros3= {
    'var_smoothing': np.logspace(0, -9, num=100)
                                             }
```

```python
rejilla3=RandomizedSearchCV(bayes,parametros3,scoring=metrica,refit='f1_macro',cv=5,n_iter=3)
```

```python
rejilla3.fit(x,y)
```

```python
mejor_score3= rejilla3.best_score_*100
mejores_parametros3=rejilla3.best_params_
```

```python
print(f' f1_macro de bayes fue : {mejor_score3 :.2f} %')
print(f'Los mejores hiper par√°metros aleatoreos de bayes fueron : {mejores_parametros3}')
```

```python
#Tabla de iteraciones de nave bayes
resultados3 =pd.DataFrame(rejilla3.cv_results_)[['params','mean_test_f1_macro','mean_test_accuracy']]
resultados3
```

```python
exactitud3=resultados3.iloc[0,2]*100
print(f'La exactitud del modelo bayes fue : {exactitud3 :.2f} %')
```

#####La exactitud con el melo de bayes fue del 75% , y un f1_macro de 42.85%, con que se concluye que el modelo no es apenas aceptable



#Tabla comparativa de los modelos

```python
modelos={
        'Algoritmo':['Arbol de decisi√≥n','knn','Nave bayes'],
        'f1_macro':[mejor_score1,mejor_score2,mejor_score3],
        'Exactitud':[exactitud1,exactitud2,exactitud3]
}
```

```python
tabla_modelos=pd.DataFrame(modelos)
tabla_modelos
```

##Para elegir el mejor modelo se tiene en cuenta tanto la exactitud, como el f1 macro,la exactitud es muy confiable cuando se tiene un set de datos balanceado, es decir, cuando hay aproximadamente el mismo porcentaje de las clases componentes de la variable objetivo, en este caso en particular hay aproximadamente 75% no contagiados y 25% contagiados,es decir el dataset est√° desbalanceado, por eso tiene mas relevancia el f1 macro ya que toma en cuenta tanto falsos positivos como falsos negativos, que es muy importante para predecir enfermedades, y aunque knn tiene una exactitud un poco mas baja de los otros modelos, tiene mejor f1 macro y en consecuencia se toma como el mejor modelo.
